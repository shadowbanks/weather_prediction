```python
import os
import zipfile
import pandas as pd

def unzip_file(file_path, destination_path):
    # Verify file exists
    if not os.path.exists(file_path):
        print(f"Error: {file_path} does not exist.")
        return

    # Ensure destination path exist, create if necessary
    os.makedirs(destination_path, exist_ok=True)

    with zipfile.ZipFile(file_path, 'r') as extract:
        extract.extractall(destination_path)
        print(f"Extracted all files to {destination_path}")

file_path = 'global-daily-climate-data.zip'
destination_path = "datasets"
# unzip_file(file_path, destination_path)
```

```python
# Load csv
cities = pd.read_csv(destination_path + '/cities.csv')
countries = pd.read_csv(destination_path + '/countries.csv')

# Load parquet
weather = pd.read_parquet(destination_path + '/daily_weather.parquet')
```
